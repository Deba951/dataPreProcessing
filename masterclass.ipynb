```python
import pandas as pd
```

```python
data = pd.read_csv("/kaggle/input/masterclass/Walmart.csv")
data
```

```python
data.dropna(how="all", inplace=True)
data
# row number 9994 has been dropped. 
#inplace=True --> permanent change.
```

```python
data ["Country"].replace({"United States" : "USA"}, inplace=True)
data
```

```python
avg = data["Sales"].mean()
```

```python
data["Sales"].fillna(value=avg, inplace=True)
data
```

```python
data.to_csv("Walmartcleandata.csv")
```
